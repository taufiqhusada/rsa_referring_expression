```python
dict_threshold_new_data_attr = defaultdict(lambda: 1)
dict_threshold_new_data_type = defaultdict(lambda: 1)

dict_threshold_old_data_attr = defaultdict(lambda: 1)
dict_threshold_old_data_type = defaultdict(lambda: 1)
```

```python
dict_avg_new_data_attr = defaultdict(lambda: 0)
dict_avg_new_data_type = defaultdict(lambda: 0)

dict_avg_old_data_attr = defaultdict(lambda: 0)
dict_avg_old_data_type = defaultdict(lambda: 0)
```

```python
list_col = df.columns
len(list_col)
```

```python
for i, row in df.iterrows():
    box_class = row['box_alias'][:row['box_alias'].find('-')]
    is_old_data = row['is_old_data']
    
    # find value type
    val_chosen = -1
    type_chosen = None
    for col in list_col:
        if ('TYPE' in col):
            val_now = row[col]
            type_now = col
            
            if (is_old_data):
                dict_avg_old_data_type[col] += val_now
            else:
                dict_avg_new_data_type[col] += val_now
            
            if (val_now > val_chosen):
                val_chosen = val_now
                type_chosen = type_now
    print(type_chosen)
    if (val_chosen != -1):        
        if (is_old_data):
            dict_threshold_old_data_type[type_chosen] = min(val_chosen, dict_threshold_old_data_type[type_chosen])
        else:
            dict_threshold_new_data_type[type_chosen] = min(val_chosen, dict_threshold_new_data_type[type_chosen])

    
    # find value attr
    val_chosen = -1
    attr_chosen = None
    for col in list_col:
        if ('ATTR' in col):
            val_now = row[col]
            attr_now = col
            
            if (is_old_data):
                dict_avg_old_data_attr[col] += val_now
            else:
                dict_avg_new_data_attr[col] += val_now
            
            if (val_now > val_chosen):
                val_chosen = val_now
                attr_chosen = attr_now
    print(attr_chosen)
    if (val_chosen != -1):        
        if (is_old_data):
            dict_threshold_old_data_attr[attr_chosen] = min(val_chosen, dict_threshold_old_data_attr[attr_chosen])
        else:
            dict_threshold_new_data_attr[attr_chosen] = min(val_chosen, dict_threshold_new_data_attr[attr_chosen])

            
for key in dict_avg_new_data_attr:
    dict_avg_new_data_attr[key] /= 18
for key in dict_avg_new_data_type:
    dict_avg_new_data_type[key] /= 18
for key in dict_avg_old_data_attr:
    dict_avg_old_data_attr[key] /= 18
for key in dict_avg_old_data_type:
    dict_avg_old_data_type[key] /= 18
```

```python
# initialize dict
dict_threshold_new_data_attr = defaultdict(lambda: 1)
dict_threshold_new_data_type = defaultdict(lambda: 1)

dict_threshold_old_data_attr = defaultdict(lambda: 1)
dict_threshold_old_data_type = defaultdict(lambda: 1)

dict_avg_new_data_attr = defaultdict(lambda: 0)
dict_avg_new_data_type = defaultdict(lambda: 0)

dict_avg_old_data_attr = defaultdict(lambda: 0)
dict_avg_old_data_type = defaultdict(lambda: 0)

for file_id in range(5000):
    if (file_id % 100 == 0):
        print(f'finish processing {file_id} files')
    df = pd.read_csv(os.path.join(data_path,f'refCOCO/test/attr_tables_with_target_box/attr_{file_id}.tsv'), encoding='utf-8',sep='\t')
    for i, row in df.iterrows():
        box_class = row['box_alias'][:row['box_alias'].find('-')]
        is_old_data = row['is_old_data']
         # find value type
        val_chosen = -1
        type_chosen = None
        for col in list_col:
            if ('TYPE' in col):
                val_now = row[col]
                type_now = col

                if (is_old_data):
                    dict_avg_old_data_type[col] += val_now
                else:
                    dict_avg_new_data_type[col] += val_now

                if (val_now > val_chosen):
                    val_chosen = val_now
                    type_chosen = type_now
#         print(type_chosen)
        if (val_chosen != -1):        
            if (is_old_data):
                dict_threshold_old_data_type[type_chosen] = min(val_chosen, dict_threshold_old_data_type[type_chosen])
            else:
                dict_threshold_new_data_type[type_chosen] = min(val_chosen, dict_threshold_new_data_type[type_chosen])


        # find value attr
        val_chosen = -1
        attr_chosen = None
        for col in list_col:
            if ('ATTR' in col):
                val_now = row[col]
                attr_now = col

                if (is_old_data):
                    dict_avg_old_data_attr[col] += val_now
                else:
                    dict_avg_new_data_attr[col] += val_now

                if (val_now > val_chosen):
                    val_chosen = val_now
                    attr_chosen = attr_now
#         print(attr_chosen)
        if (val_chosen != -1):        
            if (is_old_data):
                dict_threshold_old_data_attr[attr_chosen] = min(val_chosen, dict_threshold_old_data_attr[attr_chosen])
            else:
                dict_threshold_new_data_attr[attr_chosen] = min(val_chosen, dict_threshold_new_data_attr[attr_chosen])
```
