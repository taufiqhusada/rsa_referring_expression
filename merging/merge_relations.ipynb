```python
import json
import pandas as pd
import os
from collections import defaultdict
```

```python
old_data_path = '/scratch2/hle'
new_data_path = '/scratch2/thdaryan/data/new_result_merge_graph'
merged_data_path = '/scratch2/thdaryan/data/merged_result'
```

```python
df_old = pd.read_csv(os.path.join(old_data_path,f'refCOCO/test/rel_tables/rel_100.tsv'), encoding='utf-8',sep='\t')
df_old.head()
```

```python
df_new = pd.read_csv(os.path.join(new_data_path,f'refCOCO/test/rel_tables/rel_100.tsv'), encoding='utf-8',sep='\t')
df_new.head()
```

```python
df_merged = pd.concat([df_old, df_new])
df_merged = df_merged.reset_index(drop=True)

# check if there is NaN value (if so then there is some error on the process)
is_nan = False
for col in df_merged.columns:
    is_nan = df_merged[col].isnull().values.any()
if (is_nan):
    raise Exception('there is NaN value')
df_merged
```

```python
def merge_old_and_new_df(file_idx):
    df_old = pd.read_csv(os.path.join(old_data_path,f'refCOCO/test/rel_tables/rel_{file_idx}.tsv'), encoding='utf-8',sep='\t')
    df_new = pd.read_csv(os.path.join(new_data_path,f'refCOCO/test/rel_tables/rel_{file_idx}.tsv'), encoding='utf-8',sep='\t')
    
    df_merged = pd.concat([df_old, df_new])
    df_merged = df_merged.reset_index(drop=True)

    # check if there is NaN value (if so then there is some error on the process)
    is_nan = False
    for col in df_merged.columns:
        is_nan = df_merged[col].isnull().values.any()
    if (is_nan):
        raise Exception('there is NaN value')
    
    df_merged.to_csv(os.path.join(merged_data_path, f'refCOCO/test/rel_tables/rel_{file_idx}.tsv'), sep = '\t', index=False)
```

```python
merge_old_and_new_df(100)
```

```python
df = pd.read_csv(os.path.join(merged_data_path, f'refCOCO/test/rel_tables/rel_100.tsv'), sep = '\t')
df
```

```python
for i in range(5000):
    merge_old_and_new_df(i)
    if (i % 50 == 0 ):
        print(f'finished processing {i} files')
```
